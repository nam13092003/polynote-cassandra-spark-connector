```scala
import com.datastax.spark.connector.cql.CassandraConnector
import org.apache.spark.sql.SaveMode

CassandraConnector(spark.sparkContext.getConf).withSessionDo { session =>
  session.execute("CREATE KEYSPACE IF NOT EXISTS polynote_db WITH replication = {'class': 'SimpleStrategy', 'replication_factor': 1};")
  session.execute("CREATE TABLE IF NOT EXISTS polynote_db.users (id INT PRIMARY KEY, name TEXT, role TEXT);")
}

val data = Seq(
    (1, "Alice", "Data Engineer"), 
    (2, "Bob", "Data Scientist"),
    (3, "Charlie", "Backend Developer")
)
val df = spark.createDataFrame(data).toDF("id", "name", "role")

df.write
  .format("org.apache.spark.sql.cassandra")
  .options(Map("table" -> "users", "keyspace" -> "polynote_db"))
  .mode(SaveMode.Append)
  .save()

val readDF = spark.read
  .format("org.apache.spark.sql.cassandra")
  .options(Map("table" -> "users", "keyspace" -> "polynote_db"))
  .load()

readDF.show()
```
